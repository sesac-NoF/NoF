```python
# 코드 정리 및 빈 공백 삭제 

# 파일 경로 설정
file_path = r'C:\Users\user\Desktop\프로젝트 기획서\NoF\data\data_parties_60_prep.txt'

# 파일 읽기 및 데이터 처리
with open(file_path, 'r', encoding='utf-8') as file:
    texts = file.readlines()

datas = []
for text in texts:
    text = text.replace(' ', '')  # 공백 제거
    words = text.split(',')  # 콤마로 단어 분리
    datas.extend(words)  # 리스트에 단어 추가

# 중복을 제거하고 줄바꿈 처리
uniq_words = set(datas)
clean_words = {word.replace('\n', '') for word in uniq_words if word != '\n'}

# 최종적으로 유니크한 값만 출력
print(len(clean_words))
print(clean_words)
```

```python
# 파일 경로 설정
output_file = r'C:\Users\user\Desktop\프로젝트 기획서\NoF\data\clean_words_parties.txt'

# set 데이터를 텍스트 파일로 저장
with open(output_file, 'w', encoding='utf-8') as file:
    for word in clean_words:
        file.write(f"{word}\n")

print(f"'{output_file}' 파일로 저장되었습니다.")
```

## 1-3 불용어 중복 제거 및 정렬

```python
# 파일 경로 설정
file_path = r'C:\Users\user\Desktop\프로젝트 기획서\NoF\data\sorted_words_20241025.txt'

# 파일 읽기
with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# 단어 추출 및 정렬
words = text.split()  # 공백으로 단어 분리
unique_words = set(words)  # 중복 제거
sorted_words = sorted(unique_words)  # 오름차순 정렬

# 정렬된 단어 출력
for word in sorted_words:
    print(word)

# 정렬된 단어를 새로운 파일에 저장하려면 아래 코드 사용
with open('sorted_words.txt', 'w', encoding='utf-8') as output_file:
    for word in sorted_words:
        output_file.write(word + '\n')
```
