#### 제~조

```python
# '제~조' 만 있는 열의 값을 앞으로 이동 

import pandas as pd
import re

def move_matching_text(data3_gana):
    """
    데이터프레임의 51번 열부터 0번 열까지 역순으로 '제\d+호' 형태의 텍스트만 있는 셀을 매칭하고,
    매칭되는 경우 해당 값을 앞의 열로 이동하는 함수입니다.

    Parameters:
    data3_gana (pd.DataFrame): 텍스트를 처리할 데이터프레임

    Returns:
    pd.DataFrame: 텍스트가 이동된 데이터프레임
    """
    column_count = len(data3_gana.columns)

    for index, row in data3_gana.iterrows():
        for col in range(51, 0, -1):  # 51번 열부터 0번 열까지 역순
            if col < column_count and pd.notna(row[col]) and re.fullmatch(r'제\d+조', str(row[col]).strip()):  # '제\d+호'만 있는지 확인
                # 값을 앞의 열로 이동
                data3_gana.iat[index, col - 1] = (
                    str(data3_gana.iat[index, col - 1]) + ' ' + row[col]
                    if pd.notna(data3_gana.iat[index, col - 1])
                    else row[col]
                )
                # 이동 후 현재 열은 None으로 설정
                data3_gana.iat[index, col] = None
    return data3_gana

# # 예시 데이터프레임 생성 (52개의 열을 가정)
# data3_gana = pd.DataFrame({
#     i: [None] for i in range(52)  # 52개의 열을 생성
# })
# data3_gana.at[0, 1] = '제29호'
# data3_gana.at[0, 3] = '제28호'
# data3_gana.at[0, 51] = '근로기준법 제28호'

# 함수 사용
result_data3_gana = move_matching_text(data3_gana)

# 결과 출력
print(result_data3_gana)
```

### 전처리 : '제~' 로 시작하는 텍스트는 앞의 열로 이동

```python
import pandas as pd
import re

def move_matching_text(result_data3_gana):
    """
    데이터프레임의 마지막 열부터 0번 열까지 역순으로 '제'로 시작하는 텍스트만 있는 셀을 매칭하고,
    매칭되는 경우 해당 값을 앞의 열로 이동하거나, 앞 열에 텍스트가 있을 경우 통합하는 함수입니다.
    앞쪽에 공백이 있는 경우도 포함하여 처리합니다.

    Parameters:
    result_data3_gana (pd.DataFrame): 텍스트를 처리할 데이터프레임

    Returns:
    pd.DataFrame: 텍스트가 이동된 데이터프레임
    """
    # 데이터프레임의 총 열 개수를 계산
    column_count = len(result_data3_gana.columns)

    # 각 행을 반복하여 처리 (index는 행 번호, row는 행 데이터)
    for index, row in result_data3_gana.iterrows():
        # 마지막 열부터 0번 열까지 역순으로 순회 (열 인덱스가 범위 안에 있도록 min(column_count, 51)로 제한)
        for col in range(min(column_count - 1, 51), 0, -1):
            # 열 번호가 데이터프레임 범위 안에 있어야 함
            # 현재 셀이 비어있지 않고, 값이 '제'로 시작하는지 확인 (앞쪽에 공백이 있는 경우도 포함)
            if pd.notna(row[col]) and re.match(r'^\s*제', str(row[col]).strip()):
                # 앞의 열로 값을 이동하거나 통합
                if pd.notna(result_data3_gana.iat[index, col - 1]):
                    # 앞의 열에 값이 있는 경우 통합 (공백으로 구분하여 연결)
                    result_data3_gana.iat[index, col - 1] = str(result_data3_gana.iat[index, col - 1]) + ' ' + row[col].strip()
                else:
                    # 앞의 열이 비어있는 경우, 값을 그대로 이동
                    result_data3_gana.iat[index, col - 1] = row[col].strip()

                # 현재 열의 값을 None으로 설정하여 값이 이동되었음을 표시
                result_data3_gana.iat[index, col] = None

    return result_data3_gana

# 예시 데이터프레임 생성
# result_data3_gana = pd.DataFrame({
#     0: [None],
#     1: [' 제29조'],  # 앞쪽에 공백이 있는 경우
#     2: [None],
#     3: ['제28호'],
#     4: ['기타 내용'],
#     51: [' 제30조']  # 앞쪽에 공백이 있는 경우
# })

# 함수 사용
result_data3_gana = move_matching_text(result_data3_gana)

# 결과 출력
print(result_data3_gana)
```

### 전처리 : 앞의 열부터 텍스트 채우기

```python
import pandas as pd
import numpy as np

# 함수 정의
def organize_text_front(result_data3_gana):
    for j in range(len(result_data3_gana)):  # 각 행에 대해 반복
        # 해당 행의 NaN이 아닌 값만 리스트에 저장
        non_null_values = [result_data3_gana.iloc[j, i] for i in range(len(result_data3_gana.columns)) if pd.notna(result_data3_gana.iloc[j, i])]
        
        # 리스트의 값들을 앞에서부터 채우고 나머지는 NaN으로 설정
        for i in range(len(result_data3_gana.columns)):
            if i < len(non_null_values):
                result_data3_gana.iloc[j, i] = non_null_values[i]  # 텍스트 채우기
            else:
                result_data3_gana.iloc[j, i] = np.nan  # 나머지는 NaN으로 설정
    return result_data3_gana   

# 예시 데이터프레임 생성
# result_data3_gana = pd.DataFrame({
#     0: [None],
#     1: ['텍스트1'],
#     2: ['텍스트2'],
#     3: [None],
#     4: [None],
#     5: ['텍스트3'],
#     6: ['텍스트4']
# })

# 함수 사용
result_data3_gana = organize_text_front(result_data3_gana)

# 결과 출력
print(result_data3_gana)
```

```python
import pandas as pd
import numpy as np

# 함수 정의
def organize_text_front(result_data3_gana):
    for j in range(len(result_data3_gana)):  # 각 행에 대해 반복
        # 해당 행의 NaN이 아닌 값만 리스트에 저장
        non_null_values = [result_data3_gana.iloc[j, i] for i in range(len(result_data3_gana.columns)) if pd.notna(result_data3_gana.iloc[j, i])]
        
        # 리스트의 값들을 앞에서부터 채우고 나머지는 NaN으로 설정
        for i in range(len(result_data3_gana.columns)):
            if i < len(non_null_values):
                result_data3_gana.iloc[j, i] = non_null_values[i]  # 텍스트 채우기
            else:
                result_data3_gana.iloc[j, i] = np.nan  # 나머지는 NaN으로 설정
    return result_data3_gana   

# 예시 데이터프레임 생성
# result_data3_gana = pd.DataFrame({
#     0: [None],
#     1: ['텍스트1'],
#     2: ['텍스트2'],
#     3: [None],
#     4: [None],
#     5: ['텍스트3'],
#     6: ['텍스트4']
# })

# 함수 사용
result_data3_gana = organize_text_front(result_data3_gana)

# 결과 출력
print(result_data3_gana)
```

### 결합 : 모든 조문 통합하기
* 전처리 : 2칸 이상의 띄어쓰기 -> 1칸으로 변경, 행 문장 시작 : 빈칸 불허

```python
import pandas as pd
import numpy as np
import re

def combine_text_columns(result_data3_gana1):
    """
    데이터프레임의 각 행의 열 값을 하나로 통합하고, 각 열의 값은 슬래시('/')로 구분하여 결합합니다.
    2칸 이상의 띄어쓰기는 1칸으로 줄이고, 결과 문자열은 앞에 공백이 없도록 처리합니다.

    Parameters:
    data (pd.DataFrame): 텍스트를 처리할 데이터프레임

    Returns:
    pd.DataFrame: 'combined_text' 열에 각 행의 모든 열 값이 통합된 데이터프레임
    """
    # 각 행의 모든 열 값을 슬래시로 구분하여 결합하고 'combined_text'라는 새 열에 저장
    result_data3_gana1['combined_text'] = result_data3_gana1.apply(lambda row: '/'.join(
        str(cell).strip() for cell in row if pd.notna(cell) and cell != ''), axis=1)

    # 2칸 이상의 공백을 1칸으로 줄이고, 앞쪽 공백을 제거
    result_data3_gana1['combined_text'] = result_data3_gana1['combined_text'].apply(lambda x: re.sub(r'\s{2,}', ' ', x).strip())
    
    return result_data3_gana1

# # 예시 데이터프레임 생성 (작은 크기의 데이터프레임으로 예시)
# data = pd.DataFrame({
#     'col1': ['  텍스트1  ', None, '텍스트3'],
#     'col2': ['텍스트2', '  텍스트4  ', None],
#     'col3': [None, '텍스트5   ', '텍스트6']
# })

# 함수 사용
result_data3_gana1 = combine_text_columns(result_data3_gana1)

# 결과 출력
print(result_data3_gana1[['combined_text']])
```
