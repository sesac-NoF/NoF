```python
data['조문_분리2']=data['조문_분리2'].fillna('없음')
```

```python
# 스플릿한 값을 데이터프레임에 동적으로 추가하는 함수
def split_and_add_columns(row):
    # 각 조항을 ','로 분리하고 공백 제거
    split_values = [value.strip() for value in row.split(',')]
    
    # 스플릿된 값을 반환
    return pd.Series(split_values)

# 조문_분리2 열의 값을 스플릿하고 새로운 열 추가
split_data = data['조문_분리2'].apply(split_and_add_columns)

# 새로 생성된 열을 원래 데이터프레임에 추가
data = pd.concat([data, split_data], axis=1)

# 결과 확인
print(data)
```
