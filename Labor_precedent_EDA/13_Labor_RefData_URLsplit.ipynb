```python
import re
import pandas as pd

def process_text(data_yk_url):
    # Create a new column to store the processed text, handle null values
    data_yk_url['참조조문_전처리6'] = data_yk_url['참조조문_전처리5'].apply(lambda x: insert_law_name(x) if pd.notnull(x) else x)
    return data_yk_url

def insert_law_name(sentence):
    # Split the sentence by '/'
    sentences = sentence.split('/')
    result = []

    for s in sentences:
        # Find the first occurrence of '제~조' (e.g., 제28조, 제29조)
        match = re.search(r'(제\d+조)', s)
        if match:
            law_name = s[:match.start()].strip()  # Extract the law name before '제~조'
            law_clauses = s[match.start():].split()  # The clause part including '제~조'
            # Insert the law name before each '제~조'
            updated_sentence = ' '.join([f"{law_name} {clause}" if clause.startswith('제') else clause for clause in law_clauses])
            result.append(updated_sentence)
        else:
            result.append(s)

    # Return the reconstructed sentence with the law name inserted before '제~조'
    return '/'.join(result)

# Example dataframe
# data_yk_url = pd.DataFrame({
#     '참조조문_전처리5': ['근로기준법 제28조 제29조', '민법 제1조 제2조 제3조']
# })

# Apply the function to your dataframe
data_yk_url = process_text(data_yk_url)

print(data_yk_url)
```

```python
# Text processing function
def process_text(sentence):
    if pd.isnull(sentence):
        return sentence
    
    # Replace unwanted characters
    sentence = (
        sentence.replace('구 ', '')            # Remove '구 '
                .replace('(구)', '')          # Remove '(구)'
                .replace('(구,', '')          # Remove '(구)'
                .replace('(법률 ,', '')       # Remove '(법률 ,'
                .replace('(', '')            # Remove '('
                .replace(')', ',')           # Replace ')' with ','
                .replace('/', ',')            # Remove '/'
                .replace('제', '/제')        # Replace '제' with '/제'
    )
    
    # Split words by spaces
    words = sentence.split()
    processed_words = []
    
    for word in words:
        # Check if the word ends with '조'
        if word.endswith('조'):
            # Add a comma after '조' if it meets certain conditions
            if len(word) == 1 or word == '조 ':
                processed_words.append('조,')
            else:
                processed_words.append(f"{word},")
        else:
            processed_words.append(word)
    
    # Join processed words with spaces and return
    return ' '.join(processed_words).strip()

# Apply the function to '참조조문_전처리6' column
data_yk_url['참조조문_전처리7'] = data_yk_url['참조조문_전처리6'].apply(process_text)

# Display the result
print(data_yk_url)
```

```python
import pandas as pd

# Example DataFrame
# data_yk_url = pd.DataFrame({
#     '참조조문_전처리6': [
#         '근로기준법 제28조 구 제29조 (구)', 
#         '민법 제1조 (구) 제2조 제3조', 
#         '민법 제1조 제2조 제3조', 
#         '민법 제1조 제2조'
#     ]
# })

# Text processing function
def process_text(sentence):
    if pd.isnull(sentence):
        return sentence
    
    # Replace unwanted characters
    sentence = (
        sentence.replace('구 ', '')            # Remove '구 '
                .replace('(구)', '')          # Remove '(구)'
                .replace('(구,', '')          # Remove '(구,'
                .replace('(법률 ,', '')       # Remove '(법률 ,'
                .replace('(', '')            # Remove '('
                .replace(')', ',')           # Replace ')' with ','
                .replace('/', ',')            # Remove '/'
                .replace('제', '/제')        # Replace '제' with '/제'
    )
    
    # Split words by spaces
    words = sentence.split()
    processed_words = []
    
    for word in words:
        # Check if the word ends with '조'
        if word.endswith('조'):
            # Add a comma after '조' if it meets certain conditions
            if len(word) == 1 or word == '조 ':
                processed_words.append('조,')
            else:
                processed_words.append(f"{word},")
        else:
            processed_words.append(word)
    
    # Join processed words with spaces
    processed_sentence = ' '.join(processed_words).strip()
    
    # Remove the first comma from the end if any
    if ',' in processed_sentence[::-1]:
        processed_sentence = processed_sentence[::-1].replace(',', '', 1)[::-1]

    return processed_sentence

# Apply the function to '참조조문_전처리6' column
data_yk_url['참조조문_전처리7'] = data_yk_url['참조조문_전처리6'].apply(process_text)

# Display the result
print(data_yk_url)
```
