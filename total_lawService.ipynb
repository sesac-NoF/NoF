```python
import pandas as pd

ls_dh = pd.read_csv('data/lawService_dh2.csv')
ls_yk = pd.read_csv('data/lawService_yk.csv')
ls_js = pd.read_csv('data/lawService_js.csv')
ls_jh = pd.read_csv('data/lawService_jh.csv')
```

```python
combined_df = pd.concat([ls_dh, ls_yk, ls_js, ls_jh], ignore_index=True)
```

```python
combined_df.info()
```

```python
# 본문, 목록에서 다날리기
combined_df[combined_df['판례내용'].isnull()==True]
```

```python
null_data = combined_df[combined_df['판례내용'].isnull()==True]
null_data
```

```python
combined_df_cleaned = combined_df.dropna(subset=['사건번호'])
combined_df_cleaned
```

```python
#중복확인
duplicates = combined_df_cleaned[combined_df_cleaned.duplicated()]
duplicates
```

```python
combined_df_cleaned.to_csv('data/lawService_total.csv')
```

```python
lawsearch_total = pd.read_csv('data/lawSearch_total.csv')
```
