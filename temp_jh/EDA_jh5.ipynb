### 참조조문 시계열

```python
import pandas as pd

# CSV 파일 경로
csv_file_path1 = r'C:\Users\82102\Downloads\새 폴더 (2)\상속본문2.csv' #해당하는 경로로 변경
csv_file_path2 = r'C:\Users\82102\Downloads\새 폴더 (2)\상속목록2.csv' #해당하는 경로로 변경

# CSV 파일 불러오기
df_text = pd.read_csv(csv_file_path1)
df_list = pd.read_csv(csv_file_path2)

print(df_text.head())
```

```python
#판례일련번호기준으로 병합프레임생성하기
merged_df = pd.merge(df_list[['판례일련번호', '선고일자']], df_text[['판례정보일련번호', '참조조문']], 
                     left_on='판례일련번호',  # df_list의 칼럼명
                     right_on='판례정보일련번호'       # df_text의 칼럼명
                     )
merged_df = merged_df[['판례일련번호', '참조조문', '선고일자']]
merged_df
```

```python
merged_df['참조조문'].iloc[0]
```

```python
# 특정 기호 제거
merged_df['참조조문'] = merged_df['참조조문'].str.replace(r'\[.*?\]', '', regex=True)  # [3] 같은 기호 제거
merged_df['참조조문'] = merged_df['참조조문'].str.replace(r'/|<br/>|\n', '', regex=True)  # /, <br/> 및 \n 제거
```

```python
merged_df['참조조문'].iloc[0]
```

```python
import re
# 법률명을 추출할 새로운 열 생성
merged_df['참조법률명'] = ''

# 각 요소에 대해 처리
for index, text in merged_df['참조조문'].items():
    if isinstance(text, str):  # 문자열인지 확인
        # 정규 표현식으로 "제", "조", "항"과 숫자가 포함된 문자 제거
        cleaned_text = re.sub(r'\S*제\d+조\S*|\S*제\d+항\S*|\S*제\d+\S*', '', text)
        
        # 두개이상 공백을 콤마','로 변환
        cleaned_text = re.sub(r'\s{2,}+', ',', cleaned_text).strip(',')
        
        # 남은 법률명을 리스트로 변환하고, 공백을 제거한 후 문자열로 결합
        cleaned_law_names = [s.strip() for s in cleaned_text.split(',')]
        
        #문자열로 결합
        merged_df.at[index, '참조법률명'] = ','.join(cleaned_law_names)
merged_df
```

```python
merged_df['참조법률명'].iloc[0]
```

```python
law_list = merged_df['참조법률명'].str.cat(sep=',').split(',')

# '법' 또는 '령'이 포함된 항목 필터링
filtered_law_list = [item for item in law_list if '법' in item or '령' in item]

# 결과 출력
print(filtered_law_list)
```

```python
# 중복 제거 (처음 나온 항목만 남기기)
seen = set()
law_list = []

for law in filtered_law_list:
    if law not in seen:
        law_list.append(law)
        seen.add(law)

# 결과 출력
print(law_list)
```
