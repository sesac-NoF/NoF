### 참조조문 시계열

```python
import pandas as pd

# CSV 파일 경로
csv_file_path1 = r'C:\Users\82102\Downloads\새 폴더 (2)\상속본문2.csv' #해당하는 경로로 변경
csv_file_path2 = r'C:\Users\82102\Downloads\새 폴더 (2)\상속목록2.csv' #해당하는 경로로 변경

# CSV 파일 불러오기
df_text = pd.read_csv(csv_file_path1)
df_list = pd.read_csv(csv_file_path2)

print(df_text.head())
```

```python
#판례일련번호기준으로 병합프레임생성하기
merged_df = pd.merge(df_list[['판례일련번호', '선고일자']], df_text[['판례정보일련번호', '참조조문']], 
                     left_on='판례일련번호',  # df_list의 칼럼명
                     right_on='판례정보일련번호'       # df_text의 칼럼명
                     )
merged_df = merged_df[['판례일련번호', '참조조문', '선고일자']]
merged_df
```

```python
merged_df['참조조문'].iloc[0]
```

```python
# 특정 기호 제거
merged_df['참조조문'] = merged_df['참조조문'].str.replace(r'\[.*?\]', '', regex=True)  # [3] 같은 기호 제거
merged_df['참조조문'] = merged_df['참조조문'].str.replace(r'/|<br/>|\n', '', regex=True)  # /, <br/> 및 \n 제거
```

```python
merged_df['참조조문'].iloc[0]
```

```python
import re

# 주어진 문자열, DataFrame의 Series에서 개별 요소를 처리
results = []

for text in merged_df['참조조문']:
    if pd.isna(text):  # NaN 체크
        continue  # NaN이면 건너뜀

    # 각 텍스트를 쉼표로 나누어 처리
    for part in re.split(r',\s*', text.strip()):  # 쉼표로 나누기
        # 법률명 및 조문 번호 추출
        law_match = re.match(r'([^\s]+(?:\s+[^\s]+)*)\s+(제\d+조(?:\s+제\d+항)?)', part)
                                #법률명 추출               조문번호     항번호있으면

        if law_match:
            # 법률명과 조문 번호가 모두 존재하는 경우
            current_law, article = law_match.groups()
            results.append(f"{current_law} {article.strip()}")
        elif current_law and re.match(r'제\d+조(?:\s+제\d+항)?', part):
            # 법률명이 이전에 존재하고 조문 번호가 나오는 경우
            results.append(f"{current_law} {part.strip()}")

for result in results:
    print(result)
```

```python
# 법률명을 추출할 새로운 열 생성
merged_df['참조법률명'] = ''

# 각 요소에 대해 처리
for index, text in merged_df['참조조문'].items():
    if isinstance(text, str):  # 문자열인지 확인
        # 정규 표현식으로 "제", "조", "항"과 숫자가 포함된 문자 제거
        cleaned_text = re.sub(r'\S*제\d+조\S*|\S*제\d+항\S*|\S*제\d+\S*', '', text)
        
        # 남은 법률명을 리스트로 변환하고, 공백을 제거한 후 문자열로 결합
        cleaned_law_names = [s.strip().replace(' ', ',') for s in cleaned_text.split(',') if s.strip()]
        
        # 컴마를 포함해서 문자열로 결합
        merged_df.at[index, '참조법률명'] = ', '.join(cleaned_law_names)
merged_df
```
