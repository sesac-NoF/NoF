```python
import mysql.connector
import pandas as pd
from db_info import user,password,host,database,port

db_connection = mysql.connector.connect(
    host = host,
    user = user,
    password = password,
    database = database,
    port = port
)

cursor = db_connection.cursor()

cursor.execute("SELECT * FROM inheritance_list")

result = cursor.fetchall()

inheritance_list = pd.DataFrame(result, columns = [i[0] for i in cursor.description])

inheritance_list
```

```python
inheritance_list.info()
```

```python
# '사건번호'가 4로 시작하지 않는 행만 필터링하여 새로운 데이터프레임 생성
filtered_df = inheritance_list[~inheritance_list['사건번호'].astype(str).str.startswith('4')][['판례일련번호', '사건번호', '선고일자']]

#4로 시작하는거
df_start4 = inheritance_list[inheritance_list['사건번호'].astype(str).str.startswith('4')][['판례일련번호', '사건번호', '선고일자']]

filtered_df.info()
```

```python
df_start4
```

```python
# 사건번호에서 '다' 앞의 숫자를 추출하여 '접수년도' 컬럼 추가
filtered_df['접수년도'] = filtered_df['사건번호'].str.extract(r'(\d+)')

# 결과 확인
print(filtered_df[['사건번호', '접수년도']])
```

```python
# 접수년도가 2글자인 경우 '19'를 붙여서 4글자로 변환
filtered_df['접수년도'] = filtered_df['접수년도'].apply(
    lambda x: '19' + x if len(x) == 2 else x
)


# 결과 확인
print(filtered_df[['사건번호', '접수년도']])
```
